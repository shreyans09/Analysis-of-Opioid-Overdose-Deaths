```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("merged_deaths_factors_final.csv")
data
df_nd = data[data['State'] == 'ND']
df_wv = data[data['State'] == 'WV']
```

```python
# Summing the Provisional Drug Overdose Deaths for West Virginia (WV) and North Dakota (ND)
total_deaths_wv = df_wv['Provisional Drug Overdose Deaths'].sum()
total_deaths_nd = df_nd['Provisional Drug Overdose Deaths'].sum()

total_deaths_wv, total_deaths_nd
# Plotting the total overdose deaths in West Virginia vs North Dakota

# Data for plotting
states = ['West Virginia (WV)', 'North Dakota (ND)']
total_deaths = [total_deaths_wv, total_deaths_nd]

# Creating the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=states, y=total_deaths, palette='coolwarm')
plt.title('Total Provisional Drug Overdose Deaths in West Virginia vs North Dakota')
plt.ylabel('Total Overdose Deaths')
plt.show()
```

```python
# 6. Heatmap for Correlation Matrix
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.DataFrame(df_nd)

# Calculate totals for each education qualification
total_bachelors_or_higher = df["Bachelor's degree or higher, 2021"].sum()
total_high_school_diploma_only = df['High school diploma only, 2021'].sum()
total_less_than_high_school = df['Less than a high school diploma, 2021'].sum()

# Plot a bar chart for the state ND
plt.figure(figsize=(8, 6))

# Bar heights
totals = [total_bachelors_or_higher, total_high_school_diploma_only, total_less_than_high_school]

# Bar labels
education_labels = ['Bachelor\'s or Higher', 'High School Diploma Only', 'Less than High School']

plt.bar(education_labels, totals)
plt.title('Total Educational Attainment in ND')
plt.ylabel('Total Individuals')
plt.show()
```
